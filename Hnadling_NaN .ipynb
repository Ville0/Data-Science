7th: Finding the mean of Age column

```python
d_mean = data["Age"].mean()

print ("Mean of the Age colunm = ", d_mean)
```

8th: Now in Age column fill all the NaN values " with the mean of Age ". and calling the data

```python
data['Age'].fillna(value = data["Age"].mean(), inplace=True)
data
```
